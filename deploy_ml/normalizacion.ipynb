```python
dicc = {'STATE':[],
        'CITY': [],
        'RATING': [],
        'REVIEW_COUNT':[],
        'RATING_HOTEL':[],
        'PREDICTED_RATING_HOTEL':[]}
```

```python
with open('../deploy_ml/data_modelada.csv', 'r', encoding= 'utf-8') as f:
    next(f)
    for linea in f:
        linea = linea.replace('"', '').split(',')
        if len(linea) != 6:
            del linea[2]
        else:
            pass
        print(linea)
        dicc['STATE'].append(linea[0].replace('"', '').strip())
        dicc['CITY'].append(linea[1].replace('"', '').strip())
        dicc['RATING'].append(float(linea[2].replace('"', '').strip()))
        dicc['REVIEW_COUNT'].append(int(float(linea[3].replace('"', '').strip())))
        dicc['RATING_HOTEL'].append(float(linea[4].replace('"', '').strip()))
        dicc['PREDICTED_RATING_HOTEL'].append(float(linea[5].replace('"', '').strip()))



        
    
```

```python
for lista in dicc.values():
    print(len(lista))
```

```python
df = pd.DataFrame(dicc)
```
